```python
def FilterLimits(filter_df_in):
    all_limits = dashdataandtables.limits_table_df.copy()
    all_exp = []
    for r in range(0,len(filter_df)):
        #print(filter_df_in[r-1]['variable'])
        #print(filter_df_in[r-1]['value'])
        if filter_df_in[r-1]['data_type'] == 'number':
            filter_value = int(filter_df_in[r-1]['value'])
        else:
            filter_value = filter_df_in[r-1]['value']
        condition = all_limits[filter_df_in[r-1]['variable']]==filter_value
        limits = all_limits[condition]['expid'].tolist()
        #print(limits)
        all_exp = all_exp + limits
        #print(condition)
    all_exp = list( dict.fromkeys(all_exp) ) ## removed duplicates
    return all_exp

ae = FilterLimits(filter_df)
ae
```
